```python
import requests
from bs4 import BeautifulSoup
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import numpy as np
nltk.download('omw-1.4')
# Ensure necessary NLTK resources are available
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
# Extract HTML from the URL
url = 'https://alfaroukroute.github.io/Mealify/'
response = requests.get(url)
html_content = response.text
```

```python
print(html_content)
```

```python
# Extract Text from HTML
soup = BeautifulSoup(html_content, 'html.parser')
text = ' '.join(soup.stripped_strings)
```

```python
print(text)
```

# Preprocessing

```python
# Cleaning
cleaned_text = re.sub(r'[^a-zA-Z\s]', '', text)
```

```python
# Normalization + remove spaces
normalized_text = cleaned_text.lower().strip()
```

```python
# Tokenization
tokens = word_tokenize(normalized_text)
```

```python
# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]
```

```python
# Remove stop words
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in lemmatized_tokens if word not in stop_words]
```

```python
print(filtered_tokens)
```

```python
# Identify Unique Words
unique_words = set(filtered_tokens)
```

```python
print(f'Number of unique words: {len(unique_words)}')
```
